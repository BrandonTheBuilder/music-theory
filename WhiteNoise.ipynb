```python
import numpy as np
from IPython.display import Audio, TextDisplayObject
import matplotlib.pyplot as plt

sampleRate = 44100
length = 5
t = np.linspace(0, length, length*sampleRate)

def make_wave(frequency):
    return np.sin(2 * np.pi * frequency * t)
```

```python
# Start with a base frequency of 110Hz - This matches with concert tuning as A2
first_harmonic = make_wave(260.813)
plt.plot(t[:1000], first_harmonic[:1000])
plt.show()
Audio(first_harmonic, rate=sampleRate)
from scipy.io.wavfile import write

scaled = np.int16(first_harmonic / np.max(np.abs(first_harmonic)) * 32767)
plt.plot(t[:5000], scaled[:5000])
plt.show()
Audio(scaled, rate=sampleRate)
write('sine.wav', sampleRate, scaled)
import os
print(os.getcwd())
```

```python
def make_wave(frequency, time):
    return np.sin(2 * np.pi * frequency * time)

def make_noise(base, width, length, n=100, sampleRate = 44100):
    t = np.linspace(0, length, length*sampleRate)
    freqs = np.random.normal(loc = base, scale = width, size=n)
    def waves():
        for freq in freqs:
            wave = make_wave(freq, t)
            yield wave
    return sum(waves())
        
noise = make_noise(440, 20000, length, 5000)

plt.plot(t[:5000], noise[:5000])
plt.show()
Audio(noise, rate=sampleRate)
```

```python
from scipy.io.wavfile import write

scaled = np.int16(noise / np.max(np.abs(noise)) * 32767)
plt.plot(t[:5000], scaled[:5000])
plt.show()
Audio(scaled, rate=sampleRate)
write('whiteNoise.wav', sampleRate, scaled)
import os
print(os.getcwd())
```
